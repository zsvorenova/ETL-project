```python
import pandas as pd
import numpy as np
import requests
import json
from sqlalchemy import create_engine
```

```python
#read csv into pandas
file1 = "Resources/healthcare_census.csv"
healthcare_census = pd.read_csv(file1)
healthcare_census.head(10)
```

```python
healthcare_census = healthcare_census.drop(["Summary Level", "State FIPS"], axis=1)
healthcare_census
```

```python
healthcare_census.rename(columns={'DRG Definition': 'DRG_Definition',
                                  'Provider Id': 'Provider_Id',
                                  'Provider Name': 'Provider_Name',
                                  'Provider Street Address': 'Street_Address',
                                  'Provider City': 'City',
                                  'Provider State': 'Provider_State',
                                  'Provider Zip Code': 'Zip_Code',
                                  'Hospital Referral Region Description': 'Hospital_Referral_Region_Description',
                                  'Total Discharges': 'Total_Discharges',
                                  'Average Covered Charges': 'Average_Covered_Charges',
                                  'Average Total Payments': 'Average_Total_Payments',
                                  'Average Medicare Payments': 'Average_Medicare_Payments',
                                  'Insured Males 18-25': 'Insured_Males_18-25',
                                  'Noninsured Males 18-25': 'Noninsured_Males_18-25',
                                  'Females 18-25 with Insurance': 'Insured_Females_18-25',
                                  'Females 18-25 without insurance': 'Noninsured_Females_18-25'}, inplace=True)

list(healthcare_census.columns)
```

```python
#fill (or change) any NaNs to 0
healthcare_census.fillna(0)
```

```python
#drop any rows with no values
healthcare_census1 = healthcare_census.dropna(how="any")
healthcare_census1
```

```python
healthcare_census1 = healthcare_census1.loc[healthcare_census1["State"] == "CA", :]
healthcare_census1.set_index("Provider_Id", inplace=False)
```
